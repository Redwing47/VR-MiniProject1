```python
# ======================================
# YOLOv8 SCRATCH TRAINING ONLY
# ======================================

print("🚀 Starting YOLO Scratch Pipeline...")

# ===== INSTALL =====
!pip install ultralytics -q

# ===== IMPORT =====
from ultralytics import YOLO
import os

# ===== DATASET PATH =====
DATA_PATH = "/kaggle/input/datasets/kunalnarang47/yolo-redwing/yolo_dataset"
print(f"📁 Dataset path: {DATA_PATH}")

# ===== CREATE YAML =====
print("📝 Creating data.yaml...")

yaml_content = f"""
train: {DATA_PATH}/train/images
val: {DATA_PATH}/val/images

nc: 5

names:
  - short sleeve top
  - long sleeve top
  - shorts
  - trousers
  - skirt
"""

with open("data.yaml", "w") as f:
    f.write(yaml_content)

print("✅ data.yaml ready")


# ======================================
# 🔵 SCRATCH TRAINING
# ======================================
print("\n🔥 SCRATCH TRAINING STARTED")

model = YOLO("yolov8n.yaml")   # 🔥 FROM SCRATCH

model.train(
    data="data.yaml",
    epochs=10,
    imgsz=640,
    batch=16,
    device=0,
    project="/kaggle/working",
    name="yolo_scratch",
    verbose=False   # clean logs
)

print("✅ Scratch training completed!")


# ======================================
# 🔵 MODEL PATH
# ======================================
SCRATCH_MODEL_PATH = "/kaggle/working/yolo_scratch/weights/best.pt"

print("\n📦 Saved model:")
print("Scratch:", SCRATCH_MODEL_PATH)


# ======================================
# 🔵 VALIDATION
# ======================================
print("\n📊 VALIDATING SCRATCH MODEL...")

model = YOLO(SCRATCH_MODEL_PATH)
metrics = model.val()

print("Scratch mAP:", metrics.box.map)


# ======================================
# 🔵 SAMPLE PREDICTION
# ======================================
import random

VAL_PATH = DATA_PATH + "/val/images"
files = os.listdir(VAL_PATH)
sample = random.choice(files)

print("\n🔍 Running prediction on:", sample)

model.predict(
    source=os.path.join(VAL_PATH, sample),
    imgsz=640,
    conf=0.25,
    save=True
)

print("📂 Prediction saved in /kaggle/working/runs/detect/predict/")


# ======================================
# 🔵 DONE
# ======================================
print("\n🎉 YOLO SCRATCH PIPELINE COMPLETE")
```
